```python
from sktime.datasets import load_UCR_UEA_dataset
from sktime.classification.hybrid import HIVECOTEV2
```

```python
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
from Tools import tools

# importing data
# Read the 2D array back from the text file
dir_path = r"C:\Users\arpan\OneDrive\Documents\GRB\data\dp3-1d-7(-10_90)"
event_types = {1:'GRB',2:'TGF',3:'SGR',4:'SFLARE',5:'DISTPAR'}
event_counter = {'GRB':0,'TGF':0,'SGR':0,'SFLARE':0,'DISTPAR':0}
# Replace 'your_search_string' with the string you are looking for in file names
search_string = 'bn'
search_pattern = os.path.join(dir_path, f'*{search_string}*')
matching_files = glob.glob(search_pattern)

X = []
Y = []

for file in matching_files:
    event_type , event_name = file.split('\\')[-1].split('_')
    loaded_data = np.loadtxt(file, dtype=np.int16, delimiter='\t').astype(np.int16)
    X.append(loaded_data)
    y = [0,0,0,0,0]
    for key,value in event_types.items():
        if value in file.split('\\')[-1]:
            print(value, file)
            y[key-1] = 1
            Y.append(y)
            event_counter[value] += 1
            break
    
X = np.array(X)
Y = np.array(Y)

X = X.reshape(X.shape[0],1,X.shape[1])
print(X.shape)
print(Y.shape)

```

```python
print(event_counter)
```

```python
# Split the train and the validation set for the fitting
from sklearn.model_selection import train_test_split
X_train = (X)
Y_train = (Y)
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size = 0.3)#, random_state=2)
print("X_train shape",X_train.shape)
print("X_test shape",X_val.shape)
print("Y_train shape",Y_train.shape)
print("Y_test shape",Y_val.shape)
```

```python
hive_cote = HIVECOTEV2(
    n_jobs=-1,  # Use all available CPU cores
    stc_params={"transform_limit_in_minutes": 2}, # {"n_estimators": 10},  # reduce the number of estimators
    drcif_params={"n_estimators": 10},  # reduce the number of estimators
    arsenal_params={"n_estimators": 10},  # reduce the number of estimators
    tde_params={"n_estimators": 10},  # reduce the number of estimators
    time_limit_in_minutes=10,  # limit the time for model fitting
    save_component_probas=False, 
    verbose=1, 
    random_state=None
)
```

```python
hive_cote.fit(X_train, Y_train)
```

```python
y_pred = hive_cote.predict(X_val)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_val, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```
